```python
import pandas as pd
import numpy as np
```

```python
data = "Online Retail.xlsx"
```

```python
#Product with the lowest and highest sales in the month of December
```

```python
df = pd.read_excel(data)
df.head(20)
```

```python
df["total_sales"] = df["Quantity"] * df["UnitPrice"]
df.head()
```

```python
df = df.copy()
df.head()
```

```python
sorted_df = df.sort_values(by="total_sales", ascending = False)
sorted_df
```

```python
months = []
for i in sorted_df["InvoiceDate"].values:
    a = str(i).split("-")[1]
    months.append(a)
print(months[:30])
```

```python
sorted_df["months"] = months
sorted_df.head()
```

```python
new_df = sorted_df[sorted_df["months"] == "12"]
new_df
```

```python
new_df.tail()
```

```python
#low_sales = new_df["total_sales"] < 0]
#low_sales
```

```python
product_vals = new_df.groupby("Description")["total_sales"].sum().sort_values(ascending=True)
product_vals
```

```python
f_product = product_vals.reset_index()
f_product
```

```python
low_sales = f_product[f_product["total_sales"] < 0]
low_sales
```

```python
low_sales.head(10).plot(x="Description", y="total_sales", kind="barh")
```

```python
high_sales = f_product[f_product["total_sales"] > 0]
high_s = high_sales.sort_values(by="total_sales", ascending=False)
high_s
```

```python
high_s.head(10).plot(x="Description", y="total_sales", kind="barh")
```
